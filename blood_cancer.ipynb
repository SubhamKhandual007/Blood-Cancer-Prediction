```python
data2 = data.drop(['id','fractal_dimension_mean','texture_se','smoothness_se','symmetry_se'], axis = 1)
data2
```

```python
ip = data2.drop(['diagnosis'], axis = 1)
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.2)
```

```python
#standar scalar tarnsform
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp = alg1.predict(x_test)
yp
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1 = ConfusionMatrixDisplay(cm)
cm1.plot()
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
acc  = accuracy_score(y_test, yp)
recall = recall_score(y_test, yp)
precision = precision_score(y_test, yp)
f1 = f1_score(y_test, yp)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
alg2 = KNeighborsClassifier(n_neighbors = 5)
alg2.fit(x_train, y_train)
```

```python
yp2 = alg2.predict(x_test)
yp2
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
acc2  = accuracy_score(y_test, yp2)
recall2 = recall_score(y_test, yp2)
precision2 = precision_score(y_test, yp2)
f12 = f1_score(y_test, yp2)
```

```python
print(acc2)
print(recall2)
print(precision2)
print(f12)
```

```python
#DECISION TREE
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion = 'entropy')

#train Decision Tree Classifier
clf = clf.fit(x_train, y_train)

#predict the response for test dataset
y_pred = clf.predict(x_test)

from sklearn import metrics
accuracy5 = metrics.accuracy_score(y_test, y_pred)
recall5 = metrics.recall_score(y_test, y_pred)
print(accuracy5)
print(recall5)
```
